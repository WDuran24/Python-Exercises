```python
def tablaE1(df):
    
    #Nombre de las promociones
    df.rename(columns = {'ID_CLIENTE':'RUT_PERSONA'}, inplace = True)

    df.rename({column:column.upper() for column in df.columns}, axis=1, inplace=True)

    df1 = df.loc[(df.NOMBRE_PROMOCION.str.upper().str.contains('MYUC')==True)&(df.NOMBRE_PROMOCION.str.upper().str.contains('TRUNK')==True)&
           (df.NOMBRE_PROMOCION.str.upper().str.contains('ESCUEL')==True)&(df.NOMBRE_PROMOCION.str.upper().str.contains('DEDICA')==True)]
    df = df.drop(df1.index)

    conditions_bundle = [((~df['ID_CABLE'].isnull()) & (df['ID_INTERNET'].isnull()) & (df['ID_FONO'].isnull())),
                       ((~df['ID_CABLE'].isnull()) & (df['ID_INTERNET'].isnull()) & (~df['ID_FONO'].isnull())),
                       ((~df['ID_CABLE'].isnull()) & (~df['ID_INTERNET'].isnull()) & (~df['ID_FONO'].isnull())),
                       ((~df['ID_CABLE'].isnull()) & (~df['ID_INTERNET'].isnull()) & (df['ID_FONO'].isnull())),
                       ((df['ID_CABLE'].isnull()) & (df['ID_INTERNET'].isnull()) & (~df['ID_FONO'].isnull())),
                       ((df['ID_CABLE'].isnull()) & (~df['ID_INTERNET'].isnull()) & (~df['ID_FONO'].isnull())),
                       ((df['ID_CABLE'].isnull()) & (~df['ID_INTERNET'].isnull()) & (df['ID_FONO'].isnull()))]

    conditions_types = [(df['NOMBRE_PROMOCION'].str.upper().str.contains('NEGOCI', regex=False)==True),
                     (df['NOMBRE_PROMOCION'].str.upper().str.contains('DEDICA', regex=False)==True),
                     (df['NOMBRE_PROMOCION'].str.upper().str.contains('MYUC', regex=False)==True),
                     (df['NOMBRE_PROMOCION'].str.upper().str.contains('ESCUELAS', regex=False)==True),
                     (df['NOMBRE_PROMOCION'].str.upper().str.contains('TRUNK', regex=False)==True),
                     ((df['NOMBRE_INTERNET'].str.upper().str.contains('NEGOCI', regex=False)==True)&
                      (df['NOMBRE_CABLE'].str.upper().str.contains('NEGOCI', regex=False)==True) & 
                     (df['NOMBRE_FONO'].str.upper().str.contains('NEGOCI', regex=False)== True)),
                        (df['NOMBRE_PROMOCION'].str.upper().str.contains('GRANDES', regex=False)== True),
                        (df['NOMBRE_INTERNET'].str.upper().str.contains('ESCUEL', regex=False)== True),
                        (df['NOMBRE_INTERNET'].str.upper().str.contains('SUBTEL', regex=False)== True),
                        (df['NOMBRE_INTERNET'].str.upper().str.contains('NEGOCI', regex=False)== True),
                        (df['NOMBRE_CABLE'].str.upper().str.contains('NEGOCI', regex=False)== True),
                        (df['NOMBRE_FONO'].str.upper().str.contains('NEGOCI', regex=False)== True)]

    conditions_finix = [(df['NOMBRE_PROMOCION'].str.upper().str.contains('0322', regex=False)==True),
                       (df['NOMBRE_PROMOCION'].str.upper().str.contains('0422', regex=False)==True),
                       (df['NOMBRE_PROMOCION'].str.upper().str.contains('0522', regex=False)==True),
                       (df['NOMBRE_PROMOCION'].str.upper().str.contains('0822', regex=False)==True),
                       (df['NOMBRE_PROMOCION'].str.upper().str.contains('1122', regex=False)==True),
                       (df['NOMBRE_PROMOCION'].str.upper().str.contains('0323', regex=False)==True)]

    
    # Bundle, empresas, etc.
    choices_bundle = ['C', 'C + F', '3PLAY', 'C + I', 'F', 'I + F', 'I']
    choices_bundle2 = [1, 2, 3, 2, 1, 2, 1]
    choices_types = ['B2B','B2B','B2B','B2B','B2B','B2B','B2B','B2B','B2B','B2B','B2B','B2B']
    choices_finix = ['FINIX_MARZO','FINIX_ABRIL','VTR_MAYO','VTR_AGOSTO','VTR_NOVIEMBRE','VTR_MARZO23']
    cable =[1,1,1,1,0,0,0]
    internet=[0,0,1,1,0,1,1]
    fono=[0,1,1,0,1,1,0]

    df['TIPO'] = np.select(conditions_types, choices_types, default='B2C')
    df['BUNDLE'] = np.select(conditions_bundle, choices_bundle, default='-1')
    df['NUM_RGU'] = np.select(conditions_bundle, choices_bundle2, default= 0)
    df['FLAG_FINIX'] = np.select(conditions_finix, choices_finix, default = 'LEGACY')
    df['FLAG_CABLE'] = np.select(conditions_bundle, cable)
    df['FLAG_INTERNET'] = np.select(conditions_bundle, internet)
    df['FLAG_FONO'] = np.select(conditions_bundle, fono)

    
    #Coreccion columna de retencion y adquisicion, llave de cruce        
    ret = pd.read_csv('20230831 Descuentos Retención Activos.csv',encoding='latin-1'
                      ,on_bad_lines='skip',delimiter = ';')
    adq = pd.read_csv('20230831 Descuentos Activos Adquisición.csv', encoding='latin-1'
                      ,on_bad_lines='skip',delimiter = ';')
    per = pd.read_csv('20230831 Descuentos Permanentes Activos.csv', encoding='latin-1'
                      ,on_bad_lines='skip',delimiter = ';')
      
    ret['RUT_VIVIENDA'] = ret['id_cliente'] + ret['IDEN_VIVIENDA'].fillna(0).astype(str)
    adq['RUT_VIVIENDA'] = adq['id_cliente'] + adq['IDEN_VIVIENDA'].fillna(0).astype(str)
    per['RUT_VIVIENDA'] = per['id_cliente'] + per['IDEN_VIVIENDA'].fillna(0).astype(str)
    ret.rename({column:column.upper() for column in ret.columns}, axis=1, inplace=True)
    adq.rename({column:column.upper() for column in adq.columns}, axis=1, inplace=True)
    per.rename({column:column.upper() for column in per.columns}, axis=1, inplace=True)
    

    adq.drop_duplicates(subset = ['FECH_ALTA', 'ANOMES_ALTA', 'ID_CLIENTE','IDEN_VIVIENDA','NOMBRE_PRODUCTO', 'NOMBRE_CUENTA_SERVICIO'], inplace = True)
    ret.drop_duplicates(subset = ['FECH_ALTA', 'ANOMES_ALTA', 'ID_CLIENTE','IDEN_VIVIENDA','NOMBRE_PRODUCTO', 'NOMBRE_CUENTA_SERVICIO'], inplace = True)
    per.drop_duplicates(subset = ['FECH_ALTA', 'ANOMES_ALTA', 'ID_CLIENTE','IDEN_VIVIENDA','NOMBRE_PRODUCTO', 'NOMBRE_CUENTA_SERVICIO'], inplace = True)
    
    #Eliminar lista de descuentos que no corresponden
    
    ret['NOMBRE_PRODUCTO'] = ret['NOMBRE_PRODUCTO'].astype(str)
    ret['nuevo'] = ret['NOMBRE_PRODUCTO'].apply(contiene_L)
    
    ret['MONTO_DSCTO_LEALTAD'] = np.where(ret['nuevo'] == False, 0 , ret['MONTO_DSCTO'])
    ret['MONTO_DSCTO_NORMAL'] = np.where(ret['nuevo'] == True, 0 , ret['MONTO_DSCTO'])
    
    
    adq= pd.DataFrame(adq.groupby(['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'])['MONTO_DSCTO'].sum()).reset_index()
    adq.rename(columns = {'ID_CLIENTE':'RUT_PERSONA'}, inplace = True)

    df = pd.merge(df, adq, how = 'left', on=['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'])
    df['VALOR_DESC_COMERCIAL'] = df['MONTO_DSCTO'].values
    df.drop(columns='MONTO_DSCTO', inplace= True)
    df['VALOR_DESC_COMERCIAL'] = df['VALOR_DESC_COMERCIAL'].fillna(0)

       
    ret1= pd.DataFrame(ret.groupby(['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'])['MONTO_DSCTO_LEALTAD'].sum()).reset_index()
    ret1.rename(columns = {'ID_CLIENTE':'RUT_PERSONA'}, inplace = True)

    df = pd.merge(df, ret1, how = 'left', on=['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'])
    df['VALOR_DESC_RETENCION_LEALTAD'] = df['MONTO_DSCTO_LEALTAD'].values
    df.drop(columns='MONTO_DSCTO_LEALTAD', inplace= True)
    df['VALOR_DESC_RETENCION_LEALTAD'] = df['VALOR_DESC_RETENCION_LEALTAD'].fillna(0)
    
    ret2= pd.DataFrame(ret.groupby(['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'])['MONTO_DSCTO_NORMAL'].sum()).reset_index()
    ret2.rename(columns = {'ID_CLIENTE':'RUT_PERSONA'}, inplace = True)

    df = pd.merge(df, ret2, how = 'left', on=['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'])
    df['VALOR_DESC_RETENCION_NORMAL'] = df['MONTO_DSCTO_NORMAL'].values
    df.drop(columns='MONTO_DSCTO_NORMAL', inplace= True)
    df['VALOR_DESC_RETENCION_NORMAL'] = df['VALOR_DESC_RETENCION_NORMAL'].fillna(0)

    ret= pd.DataFrame(ret.groupby(['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'])['MONTO_DSCTO'].sum()).reset_index()
    ret.rename(columns = {'ID_CLIENTE':'RUT_PERSONA'}, inplace = True)

    df = pd.merge(df, ret, how = 'left', on=['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'])
    df['VALOR_DESC_RETENCION'] = df['MONTO_DSCTO'].values
    df.drop(columns='MONTO_DSCTO', inplace= True)
    df['VALOR_DESC_RETENCION'] = df['VALOR_DESC_RETENCION'].fillna(0)
    
    
    per= pd.DataFrame(per.groupby(['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'])['MONTO_DSCTO'].sum()).reset_index()
    per.rename(columns = {'ID_CLIENTE':'RUT_PERSONA'}, inplace = True)

    df = pd.merge(df, per, how = 'left', on=['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'])
    df['VALOR_DESC_PERMANENTE'] = df['MONTO_DSCTO'].values
    df.drop(columns='MONTO_DSCTO', inplace= True)
    df['VALOR_DESC_PERMANENTE'] = df['VALOR_DESC_PERMANENTE'].fillna(0)

    #Transformacion y suma de rentas.
    #df['RUT_VIVIENDA'] = df['RUT_PERSONA'] + df['IDEN_VIVIENDA'].fillna(0).astype(str)
    df['VALOR_RENTA_TOTAL'] = df['VALOR_RENTA_TOTAL'].fillna(0)
    df['RENTA_PRINCIPAL'] = df['VALOR_RENTA_TOTAL'] + df['VALOR_DESC_MULTISERVICIO'] 
    df['RENTA_PRINCIPAL_V2'] = df['RENTA_PRINCIPAL'] +  df['VALOR_DESC_RETENCION'] + df['VALOR_DESC_COMERCIAL'] +  df['VALOR_DESC_PERMANENTE']
    df['FLAG_CONTEO'] = 1

    for i in df.columns:
        if df[i].dtype == float:
             df[i] = df[i].fillna(0).astype(int)  
    df = df.loc[df.TIPO == 'B2C']
    df.drop_duplicates(subset = ['RUT_VIVIENDA', 'NOMBRE_CUENTA_SERVICIO','ID_PROMO'], inplace = True)
    return df
```

```python
def rangos_precios(df):
    
    df['dif_precio_1'] = df['RENTA_PRINCIPAL'] - df['PRECIO_LISTA']
    df.loc[df['dif_precio_1'] <= 0, 'dif_lista_R1'] = 'bajo_precio_lista'
    df.loc[np.logical_and(df['dif_precio_1'] > 0, df['dif_precio_1'] <= 5000), 'dif_lista_R1'] = '5mil'
    df.loc[np.logical_and(df['dif_precio_1'] > 5001, df['dif_precio_1'] <= 15000), 'dif_lista_R1'] = '5mil-15mil'
    df.loc[np.logical_and(df['dif_precio_1'] > 15001, df['dif_precio_1'] <= 25000), 'dif_lista_R1'] = '15mil-25mil'
    df.loc[df['dif_precio_1'] > 25001, 'lista'] = 'mas-25mil'

    df['dif_precio_2'] = df['RENTA_PRINCIPAL_V2'] - df['PRECIO_LISTA']
    df.loc[df['dif_precio_2'] < 0, 'dif_lista_R2'] = 'bajo_precio_lista'
    df.loc[df['dif_precio_2'] == 0, 'dif_lista_R2'] = 'igual_precio_lista'
    df.loc[np.logical_and(df['dif_precio_2'] > 0, df['dif_precio_2'] <= 5000), 'dif_lista_R2'] = '[\$1-\$5.000]'
    df.loc[np.logical_and(df['dif_precio_2'] > 5001, df['dif_precio_2'] <= 10000), 'dif_lista_R2'] = '[\$5.001-\$10.000]'
    df.loc[np.logical_and(df['dif_precio_2'] > 10001, df['dif_precio_2'] <= 15000), 'dif_lista_R2'] = '[\$10.001-\$15.000]'
    df.loc[np.logical_and(df['dif_precio_2'] > 15001, df['dif_precio_2'] <= 20000), 'dif_lista_R2'] = '[\$15.001-\$20.001]'
    df.loc[np.logical_and(df['dif_precio_2'] > 20001, df['dif_precio_2'] <= 25000), 'dif_lista_R2'] = '[\$20.001-\$25.000]'
    df.loc[np.logical_and(df['dif_precio_2'] > 25001, df['dif_precio_2'] <= 30000), 'dif_lista_R2'] = '[\$25.001-\$30.000]'
    df.loc[np.logical_and(df['dif_precio_2'] > 30001, df['dif_precio_2'] <= 35000), 'dif_lista_R2'] = '[\$30.001-\$35.000]'
    df.loc[np.logical_and(df['dif_precio_2'] > 35001, df['dif_precio_2'] <= 40000), 'dif_lista_R2'] = '[\$35.001-\$40.000]'
    df.loc[np.logical_and(df['dif_precio_2'] > 40001, df['dif_precio_2'] <= 45000), 'dif_lista_R2'] = '[\$40.001-\$45.000]'
    df.loc[df['dif_precio_2'] >= 45001, 'dif_lista_R2'] = '[+ $45.000]'

    df['dif_precio_3'] = df['RENTA_PRINCIPAL_V2'] - df['PRECIO_ADQUISICION']
    df.loc[df['dif_precio_3'] < 0, 'dif_adq_R2'] = 'bajo_precio_adquisicion'
    df.loc[df['dif_precio_3'] == 0, 'dif_adq_R2'] = 'igual_precio_adquisicion'
    df.loc[np.logical_and(df['dif_precio_3'] > 0, df['dif_precio_3'] <= 5000), 'dif_adq_R2'] = '[\$1-\$5.000]'
    df.loc[np.logical_and(df['dif_precio_3'] > 5001, df['dif_precio_3'] <= 10000), 'dif_adq_R2'] = '[\$5.001-\$10.000]'
    df.loc[np.logical_and(df['dif_precio_3'] > 10001, df['dif_precio_3'] <= 15000), 'dif_adq_R2'] = '[\$10.001-\$15.000]'
    df.loc[np.logical_and(df['dif_precio_3'] > 15001, df['dif_precio_3'] <= 20000), 'dif_adq_R2'] = '[\$15.001-\$20.001]'
    df.loc[np.logical_and(df['dif_precio_3'] > 20001, df['dif_precio_3'] <= 25000), 'dif_adq_R2'] = '[\$20.001-\$25.000]'
    df.loc[np.logical_and(df['dif_precio_3'] > 25001, df['dif_precio_3'] <= 30000), 'dif_adq_R2'] = '[\$25.001-\$30.000]'
    df.loc[np.logical_and(df['dif_precio_3'] > 30001, df['dif_precio_3'] <= 35000), 'dif_adq_R2'] = '[\$30.001-\$35.000]'
    df.loc[np.logical_and(df['dif_precio_3'] > 35001, df['dif_precio_3'] <= 40000), 'dif_adq_R2'] = '[\$35.001-\$40.000]'
    df.loc[np.logical_and(df['dif_precio_3'] > 40001, df['dif_precio_3'] <= 45000), 'dif_adq_R2'] = '[\$40.001-\$45.000]'
    df.loc[df['dif_precio_3'] >= 45001, 'dif_adq_R2'] = '[+ $45.000]'
    return df
```

```python
def velocidad(tablaE):
    tablaE.loc[tablaE['FLAG_FINIX'] == 'FINIX_ABRIL', 'FINIX_ABRIL'] = 1
    tablaE.loc[tablaE['FLAG_FINIX'] != 'FINIX_ABRIL', 'FINIX_ABRIL'] = 0

    tablaE.loc[tablaE['FLAG_FINIX'] == 'FINIX_MARZO', 'FINIX_MARZO'] = 1
    tablaE.loc[tablaE['FLAG_FINIX'] != 'FINIX_MARZO', 'FINIX_MARZO'] = 0

    tablaE.loc[tablaE['FLAG_FINIX'] == 'LEGACY', 'LEGACY'] = 1
    tablaE.loc[tablaE['FLAG_FINIX'] != 'LEGACY', 'LEGACY'] = 0

    tablaE.loc[tablaE['FLAG_FINIX'] == 'VTR_AGOSTO', 'VTR_AGOSTO'] = 1
    tablaE.loc[tablaE['FLAG_FINIX'] != 'VTR_AGOSTO', 'VTR_AGOSTO'] = 0

    tablaE.loc[tablaE['FLAG_FINIX'] == 'VTR_MARZO23', 'VTR_MARZO23'] = 1
    tablaE.loc[tablaE['FLAG_FINIX'] != 'VTR_MARZO23', 'VTR_MARZO23'] = 0

    tablaE.loc[tablaE['FLAG_FINIX'] == 'VTR_MAYO', 'VTR_MAYO'] = 1
    tablaE.loc[tablaE['FLAG_FINIX'] != 'VTR_MAYO', 'VTR_MAYO'] = 0

    tablaE.loc[tablaE['FLAG_FINIX'] == 'VTR_NOVIEMBRE', 'VTR_NOVIEMBRE'] = 1
    tablaE.loc[tablaE['FLAG_FINIX'] != 'VTR_NOVIEMBRE', 'VTR_NOVIEMBRE'] = 0
    return tablaE
```

```python
tabla_E3 = velocidad(tabla_E2)
```
