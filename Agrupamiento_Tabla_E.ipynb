```python
import pandas as pd
import datetime as dt

tabla_E = pd.read_csv((r"C:\Users\wduran\CLAROCHILE\Nicoll Constanza Aguirre Diaz - dcabello\Wilder\Revision tabla E\tabla_E3Version 1.csv"),sep=",",encoding='latin-1')
print (tabla_E)
```

```python
#Q de clientes con mas de 1 rut vivienda 
# Crear la columna 'Mas_Rut_Id_vivienda' con valores predeterminados a 0
tabla_E['Mas_Rut_Id_vivienda'] = 0

# Contar el número de 'id_vivienda' únicos para cada 'RUT_PERSONA'
conteo_id_vivienda = tabla_E.groupby('RUT_PERSONA')['RUT_VIVIENDA'].nunique()

# Identificar los registros donde hay más de un 'id_vivienda' para un 'RUT_PERSONA'
duplicados_mas_de_un_id = tabla_E['RUT_PERSONA'].isin(conteo_id_vivienda[conteo_id_vivienda > 1].index)

# Actualizar los valores de 'Mas_Rut_Id_vivienda' a 1 donde hay más de un 'id_vivienda' para un 'RUT_PERSONA'
tabla_E.loc[duplicados_mas_de_un_id, 'Mas_Rut_Id_vivienda'] = 1

# Filtrar solo los registros donde 'Mas_Rut_Id_vivienda' es igual a 1
registros_con_mas_de_un_id = tabla_E[tabla_E['Mas_Rut_Id_vivienda'] == 1]

# Guardar el resultado en un archivo CSV
registros_con_mas_de_un_id.to_csv('Registros_Con_Mas_De_Un_Id.csv', index=False)

# Imprimir la tabla actualizada
print(registros_con_mas_de_un_id)

```
