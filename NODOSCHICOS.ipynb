```python
base_filtrada = merged_df[merged_df['SUM_FLAG'] >= 1]
```

```python
# Define the bins and labels with corrected syntax
bins = [0,9001, 14001, 19001, 24001, 29001,34001,39001,44001, float('inf')]
labels = ['0-9000', '9000-14000', '14000-19000', '19000-24000', '24000-29000', '29000-34000','34000-39000','39000-44000','>44000']

# Create the new field 'RANGO_ARPU' using pd.cut with the new range logic
base_filtrada['RANGO_ARPU'] = pd.cut(base_filtrada['ARPU_FIJO_CLIENTE'], bins=bins, labels=labels, right=False)
```

```python
base_filtrada = base_filtrada[base_filtrada['BUNDLE'].str.contains('I', case=False) | (base_filtrada['BUNDLE'] == '3PLAY')]
```

```python
#ELIMINAR FINIX MARZO 
base_filtrada = base_filtrada[~((base_filtrada['SEGMENTO_CLIENTE'].str.contains('FINIX_MARZO', case=False)) & (base_filtrada['BUNDLE'] != '3PLAY'))]
```

```python
#ELIMINAR FINIX ABRIL
# Filtramos los registros para eliminar aquellos donde 'SEGMENTO_CLIENTE' incluye 'FINIX_ABRIL'
# excepto cuando 'BUNDLE' es 'C + I' o '3PLAY'

base_filtrada = base_filtrada[
    ~((base_filtrada['SEGMENTO_CLIENTE'].str.contains('FINIX_ABRIL', case=False)) & 
      (~base_filtrada['BUNDLE'].isin(['C + I', '3PLAY'])))
]
```

```python
base_filtrada = base_filtrada[
    ~((base_filtrada['SEGMENTO_CLIENTE'] == 'FINIX_ABRIL') & 
      (~base_filtrada['PROMO'].str.contains('PLUS', case=False)))
]
```
